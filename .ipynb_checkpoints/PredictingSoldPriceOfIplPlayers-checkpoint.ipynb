# Encoding x features

```python
X_features = ['AGE', 'COUNTRY', 'PLAYING ROLE',
'T-RUNS', 'T-WKTS', 'ODI-RUNS-S', 'ODI-SR-B',
'ODI-WKTS', 'ODI-SR-BL', 'CAPTAINCY EXP', 'RUNS-S',
'HS', 'AVE', 'SR-B', 'SIXERS', 'RUNS-C', 'WKTS',
'AVE-BL', 'ECON', 'SR-BL'] 

```

```python
categorical_features ={'AGE','COUNTRY','PLAYING ROLE','CAPTAINCY EXP'}
```

```python
ipl_auction_encoded_df = pd.get_dummies(ipl_auction_df[X_features],columns = categorical_features,drop_first=True)
```

```python
X_features = ipl_auction_encoded_df.columns
print(X_features)
```

# lets split the data

```python
X = sm.add_constant(ipl_auction_encoded_df)
```

```python
Y = ipl_auction_df['SOLD PRICE']
```

```python
train_X,test_X,train_Y,test_Y = train_test_split(X,Y,train_size =0.8,random_state=90)
```

# fitting the model

```python
ipl_model = sm.OLS(train_Y,train_X).fit()
```

```python
ipl_model.summary2()
```

# variace inflation factor (VIF) to check multi-collinaerity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def get_vif_factors(X):
    X_matrix = X.values
    vif = [variance_inflation_factor(X_matrix,i) for i in range(X_matrix.shape[1])]
    vif_factors = pd.DataFrame()
    vif_factors['column']= X.columns
    vif_factors['VIF']=vif
    return vif_factors
    
```

```python
vif_factors = get_vif_factors(X[X_features])
vif_factors
```

```python
columns_with_large_vif = vif_factors[vif_factors['VIF']>4].column
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(12,10))
```

```python
sn.heatmap(X[columns_with_large_vif].corr(),annot =True);
```

# Residual Analysis

```python
def draw_pp_plot(model):
    probplot = sm.ProbPlot(model.resid)
    plt.figure(figsize=(8,6))
    probplot.ppplot(line ='45')
    plt.show();
```

```python
draw_pp_plot(ipl_model_3)
```

```python
def get_std_values(vals):
    return (vals-vals.mean())/vals.std()
```

```python
plt.scatter(get_std_values(ipl_model_3.fittedvalues),get_std_values(ipl_model_3.resid))
plt.show()
```
